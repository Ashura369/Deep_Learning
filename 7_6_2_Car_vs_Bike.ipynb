```python
import keras
from keras import layers, Sequential, models
from keras.layers import Dense, Conv2D, MaxPooling2D

import warnings
warnings.filterwarnings('ignore')
```

```python
paths = "/kaggle/input/datasets/utkarshsaxenadn/car-vs-bike-classification-dataset/Car-Bike-Dataset"

train_ds = keras.utils.image_dataset_from_directory(
    paths,
    label_mode='int',
    color_mode='rgb',
    batch_size=32,
    image_size=(256, 256),
    shuffle=True,
    seed=42,
    subset='training',
    validation_split=0.2
)
val_ds = keras.utils.image_dataset_from_directory(
    paths,
    label_mode='int',
    color_mode='rgb',
    batch_size=32,
    image_size=(256, 256),
    shuffle=True,
    seed=42,
    subset='validation',
    validation_split=0.2
)
```

```python
print(f"TRAINING DATASET BATCHES : {len(train_ds)}")
print(f"VALIDATION DATASET BATCHES : {len(val_ds)}")
```

```python
import matplotlib.pyplot as plt
```

```python
for batch in train_ds.take(1):
    images, labels = batch 
    batch_size = images.shape[0] # Automatically get the size of this batch
    
    plt.figure(figsize=(20, 20))
    for i in range(batch_size):
        ax = plt.subplot(8, 4, i + 1) # Adjust rows/cols as needed
        plt.imshow(images[i].numpy().astype('uint8'))
        name = labels[i].numpy()
        plt.title(train_ds.class_names[name])
        plt.axis('off')

plt.tight_layout()
plt.show()
```
