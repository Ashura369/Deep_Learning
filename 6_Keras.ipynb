```python
# =====================================================================
# DEEP LEARNING - NATIVE KERAS IMPORTS (ERROR-FREE & STANDALONE)
# =====================================================================

# 1. Core Models & Graph Inputs
from keras import Sequential, Model, Input

# 2. Neural Network Layers (Dense, Dropout, CNNs, RNNs, and Preprocessing)
from keras.layers import (
    Dense,                 # Standard fully-connected layer
    Dropout,               # Regularization layer (prevents overfitting)
    BatchNormalization,    # Normalizes activations of the previous layer
    Flatten,               # Flattens input arrays (e.g., 2D images to 1D vectors)
    
    # Convolutional Layers (for Computer Vision)
    Conv2D, 
    MaxPooling2D, 
    
    # Recurrent Layers (for Time Series and NLP)
    SimpleRNN, 
    LSTM, 
    GRU, 
    
    # Data Preprocessing Layers (Native Scaling and Encoding)
    Normalization,         # Equivalent to scikit-learn's StandardScaler
    CategoryEncoding       # Equivalent to scikit-learn's OneHotEncoder
)

# 3. Optimizers (Equivalent to scikit-learn's Solvers)
from keras.optimizers import Adam, SGD, RMSprop, Adagrad

# 4. Loss Functions (Calculates training error)
from keras.losses import (
    MeanSquaredError,             # For standard Regression (squared_error)
    MeanAbsoluteError,            # For robust Regression (absolute_error)
    BinaryCrossentropy,           # For Binary Classification (log_loss)
    CategoricalCrossentropy,      # For Multi-class Classification (one-hot targets)
    SparseCategoricalCrossentropy # For Multi-class Classification (integer targets)
)

# 5. Performance Metrics (Monitors training quality)
from keras.metrics import (
    Accuracy, 
    BinaryAccuracy, 
    Precision, 
    Recall, 
    AUC                           # Area Under ROC Curve
)

# 6. Weight Regularizers (Equivalent to Lasso and Ridge)
from keras.regularizers import l1, l2, l1_l2

# 7. Callbacks (Controls training execution dynamically)
from keras.callbacks import (
    EarlyStopping,                # Equivalent to early_stopping=True
    ModelCheckpoint,              # Saves the best model weights during training
    ReduceLROnPlateau,            # Automatically reduces learning rate when validation stalls
    LearningRateScheduler         # Custom learning rate schedules per epoch
)

# 8. Utilities
from keras.utils import (
    to_categorical,               # Converts labels to one-hot arrays
    Sequence                      # Base class for building custom data generators
)
```
